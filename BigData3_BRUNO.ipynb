```python
#Questão 1, letra a. Agrupe os dados por tipo de bebidas;

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

bebidas = df.groupby('Tipo de Bebida')

beer = bebidas.get_group('Beer')
spirits = bebidas.get_group('Spirits')
wine = bebidas.get_group('Wine')
other = bebidas.get_group('Other')

display(beer, spirits, wine, other)
```

```python
#Questão 1, letra b. Agrupe os dados por Região e por Ano;

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

regiao_ano = df.groupby(['Região', 'Ano']).std(0)
display(regiao_ano)
```

```python
#Questão 1, letra c. Seção de Contagens: Contar a ocorrência de Regiões, de Países e a soma da coluna de valores por Bebida

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

regiao = df['Região'].nunique()
pais = df['País'].nunique()
soma_valores = df.groupby('Tipo de Bebida')['Valor'].sum()

print("Ocorrência de Regiões: ", regiao)
print("Ocorrência de Países: ", pais, "\n")
display(soma_valores)
```

```python
#Questão 1, letra d. Realize análises estatísticas da coluna dos valores: Média, Moda, Mediana, Estatística Descritiva e Gráfico de comparação dos valores agrupados por tipo de bebida.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

media = df['Valor'].mean()
mediana = df['Valor'].median()
moda = df['Valor'].mode()
estatistica = df['Valor'].describe()

print("Média: ", media, "\n")
print("Mediana: ", mediana, "\n")
print("Moda: ", moda, "\n")
print("Estatística Descritiva:", estatistica, "\n")

grafico = df.groupby('Tipo de Bebida')['Valor'].std()
display(grafico)
grafico.plot.bar(x='Tipo de Bebida', y = 'Valor')
```
