```python
#Questão 1, letra a. Agrupe os dados por tipo de bebidas;

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

bebidas = df.groupby('Tipo de Bebida')

beer = bebidas.get_group('Beer')
spirits = bebidas.get_group('Spirits')
wine = bebidas.get_group('Wine')
other = bebidas.get_group('Other')

display(beer, spirits, wine, other)
```

```python
#Questão 1, letra b. Agrupe os dados por Região e por Ano;

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

regiao_ano = df.groupby(['Região', 'Ano']).std(0)
display(regiao_ano)
```

```python
#Questão 1, letra c. Seção de Contagens: Contar a ocorrência de Regiões, de Países e a soma da coluna de valores por Bebida

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

regiao = df['Região'].nunique()
pais = df['País'].nunique()
soma_valores = df.groupby('Tipo de Bebida')['Valor'].sum()

print("Ocorrência de Regiões: ", regiao)
print("Ocorrência de Países: ", pais, "\n")
display(soma_valores)
```

```python
#Questão 1, letra d. Realize análises estatísticas da coluna dos valores: Média, Moda, Mediana, Estatística Descritiva e Gráfico de comparação dos valores agrupados por tipo de bebida.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/world_alcohol.csv', sep = ',')
df = df.rename(columns = {"Year": "Ano", "WHO region": "Região", "Country": "País", "Beverage Types": "Tipo de Bebida", "Display Value": "Valor"})
type(df)

media = df['Valor'].mean()
mediana = df['Valor'].median()
moda = df['Valor'].mode()
estatistica = df['Valor'].describe()

print("Média: ", media, "\n")
print("Mediana: ", mediana, "\n")
print("Moda: ", moda, "\n")
print("Estatística Descritiva:", estatistica, "\n")

grafico = df.groupby('Tipo de Bebida')['Valor'].std()
display(grafico)
grafico.plot.bar(x='Tipo de Bebida', y = 'Valor')
```

```python
# Questão 2, letra g. Apresente a média das mensalidades dos cursos de Medicina.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/cursos-prouni.csv', sep = ',')
type(df)

cursos = df.groupby('nome')

medicina = cursos.get_group('Medicina')
media = round(medicina['mensalidade'].mean())
print("Média das mensalidades dos cursos de Medicina:", "R$ {:.2f}".format(media))
```

```python
# Questão 2, letra h. Média das notas de corte dos cursos de tempo integral.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/cursos-prouni.csv', sep = ',')
type(df)
df['nota_integral_ampla'].fillna(0.0, inplace = True)
df['nota_integral_cotas'].fillna(0.0, inplace = True)

cursos = df.groupby('nome')

media_ampla = round(cursos['nota_integral_ampla'].mean())
print("Nota integral ampla de corte:\n", media_ampla)
print("\n")
media_cotas = round(cursos['nota_integral_cotas'].mean())
print("Notas integral cotas de corte:\n", media_cotas)
```

```python
# Questão 2, letra i. Estatística Descritiva das Notas Integral Ampla dos cursos de Bacharelado.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/bruno-jannuzzi/NF3-BIG-DATA-BrunoReis/main/cursos-prouni.csv', sep = ',')
type(df)
df['nota_integral_ampla'].fillna(0.0, inplace=True)

grau = df.groupby('grau')

bacharelado = grau.get_group('Bacharelado')
estatistica = round(bacharelado['bolsa_integral_ampla'].describe())
print("Bacharelado")
display(estatistica)
```
